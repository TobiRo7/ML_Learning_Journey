## Bar charts and heatmaps

```python
# Path of the file to read
flight_filepath = "C:/Users/T. Roth/OneDrive/Python/Machine Learning/Kaggle Projects/Data_visualization_tutorial/flight_delays.csv"

# Read the file into a variable spotify_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")

# Print the data
flight_data
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month, annot=True --> display numbers
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

```python
# Path of the file to read
insurance_filepath = "C:/Users/T. Roth/OneDrive/Python/Machine Learning/Kaggle Projects/Data_visualization_tutorial/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)

insurance_data.head()
```

```python
# display a scatterplot with indication for smoker
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
plt.xlabel('BMI')
plt.ylabel('Insurance cost in dollar')
```

```python
# add a regression line to visually see the correlation between BMI and insurance cost
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
plt.xlabel('BMI')
plt.ylabel('Insurance cost in dollar')
```

```python
# add a 2nd regression line 
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
plt.xlabel('BMI')
plt.ylabel('Insurance cost in dollar')
```

```python
# create a swarmplot which is useful for categorical data
first_300 = insurance_data[:299]
sns.swarmplot(x=first_300['smoker'],
             y=first_300['charges'])
plt.xlabel('Smoker')
plt.ylabel('Insurance cost in dollar')
```
