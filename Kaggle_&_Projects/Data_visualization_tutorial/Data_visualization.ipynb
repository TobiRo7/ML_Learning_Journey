## Bar charts and heatmaps

```python
# Path of the file to read
flight_filepath = "C:/Users/Tobias/OneDrive/Python/Machine Learning/Kaggle Projects/Data_visualization_tutorial/flight_delays.csv"

# Read the file into a variable spotify_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")

# Print the data
flight_data
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month, annot=True --> display numbers
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

## Scatter plots

```python
# Path of the file to read
insurance_filepath = "C:/Users/Tobias/OneDrive/Python/Machine Learning/Kaggle Projects/Data_visualization_tutorial/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)

insurance_data.head()
```

```python
# display a scatterplot with indication for smoker
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
plt.xlabel('BMI')
plt.ylabel('Insurance cost in dollar')
```

```python
# add a regression line to visually see the correlation between BMI and insurance cost
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
plt.xlabel('BMI')
plt.ylabel('Insurance cost in dollar')
```

```python
# add a 2nd regression line 
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
plt.xlabel('BMI')
plt.ylabel('Insurance cost in dollar')
```

```python
# create a swarmplot which is useful for categorical data
first_300 = insurance_data[:299]
sns.swarmplot(x=first_300['smoker'],
             y=first_300['charges'])
plt.xlabel('Smoker')
plt.ylabel('Insurance cost in dollar')
```

## Histograms & Density Plots

```python
# Path of the file to read
iris_filepath = "C:/Users/Tobias/OneDrive/Python/Machine Learning/Kaggle Projects/Data_visualization_tutorial/iris.csv"

# Read the file into a variable insurance_data
iris_data = pd.read_csv(iris_filepath, index_col="Id")

# Print the first 5 rows of the data
iris_data.head()
```

```python
# display petal length in a histogram
sns.histplot(iris_data['Petal Length (cm)'])
plt.ylabel('Number of iris')
```

```python
# plot the same data in a kernel density estimate plot (KDE)
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True)
```

```python
# display a 2D KDE plot with 2 kde plots and a 'hill' plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind='kde')
```

```python
# histogram for the petal length of each species
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

# Add title
plt.title("Histogram of Petal Lengths by Species")
```

```python
# KDE plots for each species
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)

# Add title
plt.title("Distribution of Petal Lengths by Species")
```

## How to choose the right diagram?
It starts with asking the right questions:
- What data do I have?
- What message do I want to transfer?

A first guideline can be:<br>
trend data over time &rarr; line plot<br>
distributed data over a range of values &rarr; density plot or histogram<br>
related data of 2 or more variables &rarr; scatter plot, bar chart or heatmap<br>
